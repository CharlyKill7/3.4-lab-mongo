# Advanced Querying Mongo

Importing libraries and setting up connection

```python
%pip install pymongo

from pymongo import MongoClient

import pandas as pd
```

```python
str_conn='mongodb://localhost:27017'

cursor=MongoClient(str_conn)

cursor
```

```python
cursor.list_database_names() 
```

```python
db = cursor.companies
colec = db.companies
```

```python
list(colec.find().limit(1))
```

### 1. All the companies whose name match 'Babelgum'. Retrieve only their `name` field.

```python
list(colec.find({'name': 'Babelgum'}, {'_id': 0, 'name': 1}))
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
list(colec.find({'number_of_employees': {'$gt': 5000}}, {'_id': 0,'name': 1, 'number_of_employees': 1}).sort('number_of_employees', -1).limit(20))
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fields.

```python
q_years={'$and': [{'founded_year': {'$lte': 2005}},
                {'founded_year': {'$gte': 2000}}]}

list(colec.find(q_years, {'name': 1, '_id': 0, 'founded_year':1}).limit(10))
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
q_value={'$and': [{'ipo.valuation_amount': {'$gt': 100000000}},
                {'founded_year': {'$lt': 2010}}]}

list(colec.find(q_value, {'_id': 0, 'name': 1, 'ipo':1}).limit(10))
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
q_2005={'$and': [{'founded_year': {'$lt': 2005}},
                {'number_of_employees': {'$lt': 1000}}]}

list(colec.find(q_2005, {'name': 1, '_id': 0, 'founded_year':1}).sort('number_of_employees', -1).limit(10))
```

### 6. All the companies that don't include the `partners` field.

```python
list(colec.find({"partners": {"$exists": False}}))
```

### 7. All the companies that have a null type of value on the `category_code` field.

```python
list(colec.find({"category_code": {"$eq": None}}).limit(2))
```

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

```python
q_emp={'$and': [{'number_of_employees': {'$gte': 100}},
                {'number_of_employees': {'$lt': 1000}}]}

list(colec.find(q_emp, {'name': 1, '_id': 0, 'number_of_employees':1}).limit(10))
```

### 9. Order all the companies by their IPO price in a descending order.

```python
list(colec.find({'ipo': {"$exists": True}} ,{'name': 1, '_id': 0, 'ipo.valuation_amount': 1}).sort('ipo.valuation_amount', -1).limit(10))
```

### 10. Retrieve the 10 companies with more employees, order by the `number of employees`

```python
list(colec.find({'number_of_employees': {"$exists": True}} ,{'name': 1, '_id': 0, 'number_of_employees': 1}).sort('number_of_employees', -1).limit(10))
```

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
q_sem={'$and': [{'founded_month': {'$gt': 6}},
                {'founded_month': {'$lte': 12}}]}

list(colec.find(q_sem, {'name': 1, '_id': 0, 'founded_month':1}).limit(10)) 
# Arriba debería poner limit(1000), pero da muchos registros y sino os voy a putear TAs!
```

### 12. All the companies founded before 2000 that have an acquisition amount of more than 10.000.00

```python
q_tenm={'$and': [{'founded_year': {'$lt': 2000}},
                {'acquisition.price_amount': {'$gt': 10000000}}]}

list(colec.find(q_tenm, {'name': 1, '_id': 0, 'founded_year':1, 'acquisition.price_amount':1}).limit(10)) 
```

### 13. All the companies that have been acquired after 2010, order by the acquisition amount, and retrieve only their `name` and `acquisition` field.

```python
q_acq={'acquisition.acquired_year': {'$gt': 2010}}

list(colec.find(q_acq,{'name': 1, '_id': 0, 'acquisition':1}).sort('acquisition.price_amount', -1).limit(10)) 
```

### 14. Order the companies by their `founded year`, retrieving only their `name` and `founded year`.

```python
list(colec.find({'founded_year': {"$exists": True}} ,{'name': 1, '_id': 0, 'founded_year': 1}).sort('founded_year', -1).limit(50))
```

### 15. All the companies that have been founded on the first seven days of the month, including the seventh. Sort them by their `acquisition price` in a descending order. Limit the search to 10 documents.

```python
q_7_days={'$and': [{'founded_day': {'$gte': 1}},
                {'founded_day': {'$lte': 7}}]}

list(colec.find(q_7_days, {'name': 1, '_id': 0, 'founded_day':1, 'acquisition.price_amount':1}).sort('acquisition.price_amount', -1).limit(10)) 
```

### 16. All the companies on the 'web' `category` that have more than 4000 employees. Sort them by the amount of employees in ascending order.

```python
q_web={'$and': [{'category_code': {'$eq': 'web'}},
                {'number_of_employees': {'$gt': 4000}}]}

list(colec.find(q_web, {'name': 1, '_id': 0, 'category_code':1, 'number_of_employees':1}).sort('number_of_employees', 1).limit(10)) 
```

### 17. All the companies whose acquisition amount is more than 10.000.000, and currency is 'EUR'.

```python
q_eur={'$and': [{'acquisition.price_currency_code': {'$eq': 'EUR'}},
                {'acquisition.price_amount': {'$gt': 10000000}}]}

list(colec.find(q_eur, {'name': 1, '_id': 0, 'acquisition.price_currency_code':1, 'acquisition.price_amount':1}).limit(10)) 
```

### 18. All the companies that have been acquired on the first trimester of the year. Limit the search to 10 companies, and retrieve only their `name` and `acquisition` fields.

```python
q_1_tm={'$and': [{'founded_month': {'$gte': 1}},
                {'founded_month': {'$lt': 4}}]}

list(colec.find(q_1_tm, {'name': 1, '_id': 0, 'acquisition':1}).limit(10)) 
```

# Bonus
### 19. All the companies that have been founded between 2000 and 2010, but have not been acquired before 2011.

```python
q_f_a={'$and': [{'founded_year': {'$gte': 2000}},
                {'founded_year': {'$lte': 2010}},
               {'$or':[{'acquisition.acquired_year':  {'$gt': 2011}},
               {'acquisition.acquired_year': {'$exists': False}}]}]}

list(colec.find(q_f_a, {'name': 1, '_id': 0, 'acquisition.acquired_year':1, 'founded_year': 1}).limit(10)) 
```

### 20. All the companies that have been 'deadpooled' after the third year.

```python
q_dead={'$and': [{'founded_year': {'$ne': None}},
                {'$expr': {'$gt': ['$deadpooled_year', {'$add': ['$founded_year', 3]}]}}]}

list(colec.find(q_dead, {'name': 1, '_id': 0, 'deadpooled_year':1, 'founded_year':1}))
```
